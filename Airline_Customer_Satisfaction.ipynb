```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"C:\Users\newyo\Downloads\Airline_customer_satisfaction.csv\Airline_customer_satisfaction.csv")
```

```python
df.head()
```

```python
# Set the aesthetic style of the plots
sns.set_style("whitegrid")
```

```python
# Visualization 1: Customer Satisfaction Distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='satisfaction')
plt.title('Customer Satisfaction Distribution')
plt.xlabel('Satisfaction')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 2: Customer Type Distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Customer Type')
plt.title('Customer Type Distribution')
plt.xlabel('Customer Type')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 3: Age Distribution
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Age', bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 4: Type of Travel Distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Type of Travel')
plt.title('Type of Travel Distribution')
plt.xlabel('Type of Travel')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 5: Class Distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Class')
plt.title('Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
# Convert categorical columns to numeric values
df['satisfaction'] = df['satisfaction'].map({'satisfied': 1, 'neutral': 0, 'dissatisfied': -1})
df['Customer Type'] = df['Customer Type'].map({'Loyal Customer': 1, 'disloyal': 0})  # Adjust the mapping as per your data
df['Type of Travel'] = df['Type of Travel'].map({'Personal Travel': 0, 'Business travel': 1})  # Adjust the mapping as per your data
df['Class'] = df['Class'].map({'Eco': 0, 'Business': 1})  # Adjust the mapping as per your data

# Visualization 6: Correlation Matrix
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Visualization 7: Flight Distance Distribution
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Flight Distance', bins=30, kde=True)
plt.title('Flight Distance Distribution')
plt.xlabel('Flight Distance')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 8: Departure Delay vs Arrival Delay
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Departure Delay in Minutes', y='Arrival Delay in Minutes')
plt.title('Departure Delay vs Arrival Delay')
plt.xlabel('Departure Delay (minutes)')
plt.ylabel('Arrival Delay (minutes)')
plt.show()
```

```python
# Visualization 9: Boxplot of Service Ratings
service_cols = ['Seat comfort', 'Food and drink', 'Inflight wifi service', 'Inflight entertainment',
                'Online support', 'On-board service', 'Leg room service', 'Baggage handling', 'Checkin service',
                'Cleanliness', 'Online boarding']
plt.figure(figsize=(12, 8))
sns.boxplot(data=df[service_cols])
plt.title('Boxplot of Service Ratings')
plt.xlabel('Service')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualization 10: Average Service Ratings by Class
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Class', y='Seat comfort', estimator=np.mean)
plt.title('Average Seat Comfort by Class')
plt.xlabel('Class')
plt.ylabel('Average Seat Comfort')
plt.show()
```
